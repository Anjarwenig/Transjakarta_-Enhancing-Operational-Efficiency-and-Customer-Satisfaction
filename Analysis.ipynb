 <b><span style='color:darkblue'>Distance</span></b>

Creating a column that calculates the distance between tapInStops and tapOutStops using latitude and longitude coordinates can provide useful insights on travel patterns and trip lengths, as well as potentially give insights in route optimization and service reliability. Distances can be calculated using coordinates to better understand geographical trends and riders' spatial behavior.

```python
# Function to calculate Haversine distance
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    a = np.sin(delta_phi / 2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2)**2
    return R * 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a)) * 1000  # for meters

df['distance'] = df.apply(lambda x: haversine(x['tapInStopsLat'], x['tapInStopsLon'], x['tapOutStopsLat'], x['tapOutStopsLon']), axis=1)

df['distance'] = np.floor(df['distance'])


# check the new columns
print(df[['distance']].head(2))
```

 <b><span style='color:darkblue'>TravelTime</span></b>

Creating travelTime column help to study how long travels take between points, which is essential for evaluating the transit system's efficiency and efficacy. 
Calculating journey time also paves the way for subsequent investigations, such as predicting traffic speeds and assessing service reliability.

```python
# Calculate travel time in minutes
df['travelTime'] = (df['tapOutTime'] - df['tapInTime']).dt.total_seconds() / 60
df['travelTime'] = np.floor(df['travelTime'])

# check the new columns
print(df[['travelTime']].head(2))
```
